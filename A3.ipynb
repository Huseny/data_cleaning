```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

#### Load the data

```python
data = pd.read_csv("outliers_handled.csv")
```

# Step 1: Normalization

#### Before normalization

```python
data.describe()
```

#### Define the scaler and Fit and trandform Data

```python
scaler = MinMaxScaler()
```

```python
data_normalized = pd.DataFrame(scaler.fit_transform(data.drop('student_id', axis=1)), columns=data.columns[1:])
```

### Bring back the non-normalized column

```python
data_normalized.insert(0, 'student_id', data['student_id'])
```

#### After Normalization

```python
data_normalized.describe()
```

```python
data.to_csv("normalized_data.csv", index=False)
```
